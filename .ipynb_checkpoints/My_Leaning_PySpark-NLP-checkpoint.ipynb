````python
stringindexer = StringIndexer(inputCol='sentiment', outputCol='label')
````

````python
data_base = stringindexer.fit(data_base).transform(data_base)
````

````python
data_base\
    .groupBy(['sentiment','label'])\
    .count()\
    .show()
````

````python
tokenizer = Tokenizer(inputCol='clear_text', outputCol='tokens')

stopwords = StopWordsRemover(inputCol='tokens', outputCol='final_text')

hashingTF = HashingTF(inputCol=stopwords.getOutputCol(), outputCol='HTF', numFeatures=1000)

tfidf = IDF(inputCol='HTF', outputCol='features')

pipeline = Pipeline(stages=[tokenizer, stopwords, hashingTF, tfidf])
````

````python
transform_data = pipeline.fit(data_base).transform(data_base)
````

````python
transform_data.show(n=2, vertical=True)
````

### Decison Tree Classifier

````python
featureDF = transform_data\
    .select(
        [
            'features',
            'label'
        ]
    )
````

````python
featureDF.show(n=5)
````

````python
featureDF.printSchema()
````

````python
tokenizer = Tokenizer(inputCol='clear_text', outputCol='tokens')

stopwords = StopWordsRemover(inputCol='tokens', outputCol='final_text')

hashingTF = HashingTF(inputCol=stopwords.getOutputCol(), outputCol='HTF', numFeatures=1000)

tfidf = IDF(inputCol='HTF', outputCol='features')

dtc = DecisionTreeClassifier(featuresCol='features', labelCol='label', maxDepth=10)

pipeline = Pipeline(stages=[tokenizer, stopwords, hashingTF, tfidf, dtc])
````

### Sample And Training

````python
training, test = data_base.randomSplit([0.70, 0.30], seed=33)
````

````python
model_dtc_training = pipeline.fit(training)
````

````python
predict = model_dtc_training.transform(test)
````

````python
predict.limit(10).toPandas()
````

### Metrics

````python
predict\
    .select(
        [
            'label',
            'prediction'
        ]
    )\
    .show()
````

````python
evaluator = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='accuracy')
````

````python
accuracy = evaluator.evaluate(predict)
````

````python
print('Accuracy = %s' % (accuracy))
````
