#### Removing Special Characters

```python
database = database.withColumn('text_regex', f.regexp_replace('text_en', "[\$#,\"!%&'()*+-./;:<=>?@^_`´{|}~\\\\]", ''))
```

```python
database.show()
```

```python
database\
    .select(
        'text_en',
        'text_regex'
    )\
    .show(n=2, truncate=False, vertical=True)
```

#### Removing Whitespace

```python
database = database.withColumn('text_clear', f.trim(database.text_regex))
```

```python
database\
    .select(
        'text_en',
        'text_clear'
    )\
    .show()
```

```python
database\
    .select(
        'text_en',
        'text_clear'
    )\
    .show(n=2, truncate=False, vertical=True)
```

### Tokenization

```python
tokenizer = Tokenizer(inputCol='text_clear', outputCol='tokens')
```

```python
tokenized = tokenizer.transform(database)
```

```python
tokenized\
    .select(
        'text_clear',
        'tokens'
    )\
    .show(n=10)
```

#### Counting The Tokens

```python
countTokens =  f.udf(lambda tokens: len(tokens), IntegerType())
```

```python
tokenized\
    .select(
        'text_clear',
        'tokens'
    )\
    .withColumn(
        'count_tokens',countTokens(f.col('tokens'))
    )\
    .show(n=10)
```

### StopWords

#### Removing Stopwords

```python
stopwords_pt = StopWordsRemover.loadDefaultStopWords('portuguese')
```

```python
stopwords = StopWordsRemover(inputCol='tokens', outputCol='text_final')
```

```python
feature_data = stopwords.transform(tokenized)
```

```python
feature_data.show(n=5, vertical=True)
```

#### Counting Clean Text Tokens

```python
feature_data\
    .select(
        'tokens',
        'text_final'
    )\
    .withColumn('count_tokens', countTokens(f.col('tokens')))\
    .withColumn('count_clear_tokens', countTokens(f.col('text_final')))\
    .show(n=10)
```

### HashingTF

```python
hashingTF = HashingTF(inputCol='text_final', outputCol='text_HTF', numFeatures=1000)
```

```python
data_HTF = hashingTF.transform(feature_data)
```

```python
data_HTF\
    .select(
        'text_final',
        'text_HTF'
    )\
    .show(n=2, truncate=False, vertical=True)
```

### TF-IDF

```python
idf = IDF(inputCol='text_HTF', outputCol='features')
```

```python
model_idf= idf.fit(data_HTF)
```

```python
idf_data = model_idf.transform(data_HTF)
```

```python
idf_data\
    .select(
        'text_final',
        'features'
    )\
    .show(n=2, truncate=False, vertical=True)
```

### Coding These Variables Transforming Them Into Numbers

```python
stringindexer = StringIndexer(inputCol='sentiment', outputCol='label')
```

```python
database = stringindexer.fit(database).transform(database)
```

### Decision Tree Classifier

```python
dtc = DecisionTreeClassifier(featuresCol='features', labelCol='label',maxDepth=10,seed=33,maxBins=10)
```

### Assembling Pipeline

```python
tokenizer = Tokenizer(inputCol='text_clear', outputCol='tokens')

stopwords = StopWordsRemover(inputCol='tokens', outputCol='text_final')

hashingTF = HashingTF(inputCol='text_final', outputCol='text_HTF', numFeatures=5000)

idf = IDF(inputCol='text_HTF', outputCol='features')

dtc = DecisionTreeClassifier(featuresCol='features', labelCol='label',maxDepth=10,maxBins=10, seed=33)

pipeline_dtc = Pipeline(stages=[tokenizer, stopwords, hashingTF, idf, dtc])
```

### Training 

#### Dividing Training And Test Base

```python
training, test = database.randomSplit([0.70, 0.30], seed=33)
```

```python
training.count()
```

```python
test.count()
```

#### Training Model

```python
model_dtc = pipeline_dtc.fit(training)
```

```python
prediction_dtc = model_dtc.transform(test)
```

```python
prediction_dtc.limit(10).toPandas()
```

### Metrics

```python
prediction_dtc\
    .select(
        'label',
        'prediction'
    )\
    .show(n=10, truncate=False)
```

```python
evaluator = MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='accuracy')
```

```python
accuracy = evaluator.evaluate(prediction_dtc)
```

```python
print('='*30)
print('Accuracy = %s' % (accuracy))
print('='*30)
```
